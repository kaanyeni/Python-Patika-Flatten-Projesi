```python

def flatten (x):
    for i in x:
        if type (i) != list:
            l.append(i)
        elif type (i) == list:
            for t in i:
                if type (t) != list:
                    l.append(t)
                elif type (t) == list:
                    flatten (t)
    return l                  
input =  [[1,'a',['cat'],2],[[[3]],'dog'],4,5]
l = []
l = flatten(input)
print (l)
```
